```python
import pylab as plt
import numpy as np
import scipy.integrate as ingrt
%matplotlib inline
```

```python
#Parameters

N=4
p=0.5
eta=2*3.1428/(N*p)

mu=0.88
V0=500
muV0=mu*V0

E=200
muE=mu*E

Omega=50
```

```python
#Velocity function for eq. of motion

def velocity(a,t):
    x,y=a
    return [muV0*eta**2*np.exp(-np.abs(eta*y))*np.cos(x)-Omega,
            muV0*eta*np.exp(-np.abs(eta*y))*np.sin(x)-muE]
```

```python
#Generate ion trajectories by solving E.O.M.

t=np.linspace(0,1,10000)

sols=[]
for phi in np.linspace(0,6.28,30):
    sols.append(ingrt.odeint(velocity,[phi,10],t))
```

```python
#Make x y plot

plt.figure(figsize=(5,5),dpi=200)
for s in sols:
    plt.plot((np.array(s[:,0])+Omega*t)/eta,np.array(s[:,1]))
plt.ylim(-3,3)
plt.xlim(-2,8)
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.title("V = " + str(V0)+ "V, E = "+str(E)+"V/cm, p = "+str(p) +"cm, $\Omega$ = "+str(Omega)+"Hz" )
plt.plot([-20,20],[0,0],'--',color='black',linewidth=1)
plt.xlim(-1,4)
```

```python
#Make phi y plot

plt.figure(figsize=(5,5),dpi=200)
for s in sols:
    plt.plot(np.array(s[:,0]),np.array(s[:,1]))
plt.ylim(-10,10)
plt.xlim(-5,5)
plt.xlabel("$\phi$")
plt.ylabel("$y$")
plt.title("V = " + str(V0)+ "V, E = "+str(E)+"V/cm, p = "+str(p) +"cm, $\Omega$ = "+str(Omega)+"Hz" )
plt.plot([-20,20],[0,0],'--',color='black',linewidth=1)
```
